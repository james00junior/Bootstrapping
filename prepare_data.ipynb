```python
def load_data2_corpus(fasta_address, label_modifying_func=str):
        '''
        :param fasta_address:
        :param label_modifying_func: extract label from the header
        '''
        labels=[]
        corpus=[]
        for cur_record in SeqIO.parse(fasta_address, 'fasta'):
            corpus.append(str(cur_record.seq).lower())
            x = str(cur_record).lower().split('\n')
            label = x[2].split()[7].split('=')
            labels.append(label[1])
        return corpus, labels
```
